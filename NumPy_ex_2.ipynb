На складі зберігаються мобільні телефони iPhone 6, iPhone 11, iPhone 12 моделей. Всього **1328** екземплярів. Моделей iPhone 6 на **120 менше**, ніж моделей iPhone 11, та на **100 більше**, ніж моделей iPhone 12. Скільки моделей кожного виду на складі?

Нехай кількість iPhone6 == x; iPhone11 == y; iPhone12 == z

тоді систему рівнянь можна записати наступним чином:

1 * x + 1 * y + 1 * z = 1328

-1 * x + 1 * y + 0 * z = 120

1 * x + 0 * y - 1 * z = 100

```python
import numpy as np
A = np.array([[1, 1, 1], [-1, 1, 0], [1, 0, -1]])
B = np.array([1328, 120, 100])
```

```python
x = np.linalg.solve(A, B)
```

```python
x[0]
```
