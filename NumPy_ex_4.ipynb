Складіть рівняння параболи що проходить через задані три точки (1, 12), (3, 54), (-1, 2). Необхідно знайти коефіцієнти рівняння a,b,c

y = a * x ^ 2 + b * x + c

Запишемо систему лінійних рівнянь

12 = 1 ^ 2 * a + 1 * b + c

54 = 3 ^ 2 * a + 3 * b + c

2 = 1 ^ 2 * a - 1 * b + c

Складемо матрицю і визначемо коефіцієнти a, b, c

```python
import numpy as np
A = np.matrix([[1, 1, 1], [9, 3, 1], [1, -1, 1]])
B = np.array([12, 54, 2])
```

```python
x = np.linalg.solve(A, B)
print(f"a = {x[0]}, b = {x[1]},  c = {x[2]}")
```
