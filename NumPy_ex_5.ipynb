Складіть функцію get_polynom, яка за набором точок виду 
(x1, y1), (x2, y2), ...... , (x1+n, y1+n) буде знаходити коефіцієнти багаточлена ci. Таким чином функція приймає список кортежів з координатами та повертає набір коефіцієнтів c0, c1 ..... cn.

```python
import numpy as np
```

```python
def get_polynom(coords):

    A_no_matrix = []
    B_no_matrix = []

    for point in coords:
        A_no_matrix.append([point[0]**2, point[0], 1])
        B_no_matrix.append(point[1])
    
    A = np.array(A_no_matrix)
    B = np.array(B_no_matrix)

    return np.linalg.solve(A, B)
```

```python
variables = get_polynom([(1, 12), (3, 54), (-1, 2)])
```
