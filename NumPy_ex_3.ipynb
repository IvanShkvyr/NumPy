Складіть канонічне рівняння еліпсоїда, що проходить через три точки A, B, C

A(3^1/2;0;3^1/2), B(6^1/2;1/2;0),(1;1/(3^1/2);1)


Канонічне рівняння еліпсоїда у прямокутній декартовій системі координат має вигляд:


x^2/a^2+y^2/b^2+z^2/c^2 = 1
​

Завдання полягає у визначенні параметрів 
a^2, b^2, c^2

Нехай 1/a^2 = c: 1/b^2 = d; 1/c^2 = i
тоді можна записати систему рівнянь:

3 * с + 0 * d + 3 * i = 1

6 * c + 1/4 * d + 0 * i = 1

1 * c + 1/3 * d + 1 * i = 1

перемножаємо третє рівняння на 3, а друге на 4 і отримуємо матрицю:

```python
import numpy as np
A = np.matrix([[3, 0, 3], [24, 1, 0], [3, 1, 3]])
B = np.array([1, 4, 3])
```

```python
x = np.linalg.solve(A, B)
print(f"a = +-{np.sqrt(1/x[0])}, b = +-{np.sqrt(1/x[1])},  c = +-{np.sqrt(1/x[2])}")
```
